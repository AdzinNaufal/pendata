###  3. Ambil contoh threshold split (misalnya):

* Threshold = 171 cm (nilai tengah antara 170 dan 172)

Maka kita akan membagi:

* Group Left (≤ 171)
* Group Right (> 171)

###  4. Hitung Entropy Awal (sebelum split)

$$Entropy(D) = -p_1 \log_2(p_1) - p_2 \log_2(p_2)$$

$$p_1 = \frac{8}{15},\quad p_2 = \frac{7}{15}$$

$$Entropy(D) = -\frac{8}{15} \log_2 \frac{8}{15} - \frac{7}{15} \log_2 \frac{7}{15} \approx 0.996$$

###  5. Entropy setelah split

Group Left (≤ 171):

* Data: No 12, 10, 15, 8, 2, 6, 11, 4, 14, 9 → **10 data**
* Kelas: 7 Homo erectus, 3 Homo sapiens

$$Entropy(\text{left}) = -\frac{7}{10}\log_2\frac{7}{10} - \frac{3}{10}\log_2\frac{3}{10} \approx 0.881$$

Group Right (> 171):

* Data: No 5, 1, 13, 7, 3 → **5 data**
* Kelas: 1 Homo erectus, 4 Homo sapiens

$$Entropy(\text{right}) = -\frac{1}{5} \log_2 \frac{1}{5} - \frac{4}{5} \log_2 \frac{4}{5} \approx 0.722$$

Information Gain:

$$IG = Entropy(D) - \left( \frac{10}{15} \cdot 0.881 + \frac{5}{15} \cdot 0.722 \right) \approx 0.996 - (0.587 + 0.241) \approx \boxed{0.168}$$

###  6. Gini Index

$$Gini(D) = 1 - \sum p_i^2 = 1 - \left(\frac{8}{15}\right)^2 - \left(\frac{7}{15}\right)^2 \approx 0.497$$

Group Left (≤ 171):

* 7 Homo erectus, 3 Homo sapiens

$$Gini(\text{left}) = 1 - \left(\frac{7}{10}\right)^2 - \left(\frac{3}{10}\right)^2 = 1 - 0.49 - 0.09 = 0.42$$

Group Right (> 171):

* 1 Homo erectus, 4 Homo sapiens

$$Gini(\text{right}) = 1 - \left(\frac{1}{5}\right)^2 - \left(\frac{4}{5}\right)^2 = 1 - 0.04 - 0.64 = 0.32$$

Gini Split:

$$Gini_{split} = \frac{10}{15} \cdot 0.42 + \frac{5}{15} \cdot 0.32 \approx 0.28 + 0.107 \approx \boxed{0.387}$$

## Visualisasi Decision Tree

Ini adalah code program untuk mencari decision tree sekaligus rules

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.tree import DecisionTreeClassifier, plot_tree, export_text
from sklearn.preprocessing import LabelEncoder

# === Data ===
data = {
    "Tinggi Badan (cm)": [175, 162, 180, 168, 172, 165, 178, 160, 170, 155, 165, 150, 175, 168, 158],
    "Berat Badan (kg)": [78, 65, 82, 70, 75, 68, 70, 58, 62, 55, 60, 50, 65, 61, 57],
    "Kelas": [
        "Homo Sapiens", "Homo Sapiens", "Homo Sapiens", "Homo Sapiens", "Homo Sapiens",
        "Homo Sapiens", "Homo Sapiens", "Homo Erectus", "Homo Erectus", "Homo Erectus",
        "Homo Erectus", "Homo Erectus", "Homo Erectus", "Homo Erectus", "Homo Erectus"
    ]
}
df = pd.DataFrame(data)

# === Label Encoding ===
le = LabelEncoder()
df['Label'] = le.fit_transform(df['Kelas'])  # tigris=1, leo=0

# === Fitur dan Target ===
X = df[["Tinggi Badan (cm)", "Berat Badan (kg)"]]
y = df["Label"]  # <- gunakan label numerik hasil encoding

# === Train Decision Tree ===
clf = DecisionTreeClassifier(criterion="gini", max_depth=3, random_state=42)
clf.fit(X, y)

# === 1. Tampilkan Pohon Keputusan ===
print("\n=== Pohon Keputusan (Decision Tree) ===")
tree_rules = export_text(clf, feature_names=list(X.columns))
print(tree_rules)
```

## Visualisasi Splitting

```python
# === Uji 2 Data Baru ===
uji_baru = pd.DataFrame({
    'Tinggi Badan (cm)': [173, 177],
    'Berat Badan (kg)': [57, 68]
})

prediksi = clf.predict(uji_baru)
hasil = le.inverse_transform(prediksi)

print("\n=== Prediksi 2 Data Baru ===")
for i, (p, b) in enumerate(zip(uji_baru['Tinggi Badan (cm)'], uji_baru['Berat Badan (kg)'])):
    print(f"Data {i+1}: Tinggi={p}, Berat={b} → Prediksi: {hasil[i]}")

# === Visualisasi Decision Boundary ===
print("\n=== Visualisasi Decision Boundary ===")
plot_colors = "ry"
plot_step = 1.0

x_min, x_max = X['Tinggi Badan (cm)'].min() - 10, X['Tinggi Badan (cm)'].max() + 10
y_min, y_max = X['Berat Badan (kg)'].min() - 10, X['Berat Badan (kg)'].max() + 10

xx, yy = np.meshgrid(np.arange(x_min, x_max, plot_step),
                     np.arange(y_min, y_max, plot_step))

grid = np.c_[xx.ravel(), yy.ravel()]
grid_df = pd.DataFrame(grid, columns=['Tinggi Badan (cm)', 'Berat Badan (kg)'])
Z = clf.predict(grid_df)
Z = Z.reshape(xx.shape)

plt.figure(figsize=(8, 6))
plt.contourf(xx, yy, Z, cmap=plt.cm.RdYlBu, alpha=0.6)

# Plot titik data training
for label, color in zip([0, 1], plot_colors):
    idx = np.where(y == label)
    plt.scatter(X.iloc[idx]['Tinggi Badan (cm)'], X.iloc[idx]['Berat Badan (kg)'],
                c=color, edgecolor='k', label=le.inverse_transform([label])[0])

plt.xlabel("Tinggi Badan (cm)")
plt.ylabel("Berat Berat (kg)")
plt.title("Decision Boundary (Tinggi vs Berat)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Visualisasi Splitting Setelah ada 2 Data Baru

```python
# === Visualisasi Decision Boundary ===
print("\n=== Visualisasi Decision Boundary ===")
plot_colors = "ry"
plot_step = 1.0

x_min, x_max = X['Tinggi Badan (cm)'].min() - 10, X['Tinggi Badan (cm)'].max() + 10
y_min, y_max = X['Berat Badan (kg)'].min() - 10, X['Berat Badan (kg)'].max() + 10

xx, yy = np.meshgrid(np.arange(x_min, x_max, plot_step),
                     np.arange(y_min, y_max, plot_step))

grid = np.c_[xx.ravel(), yy.ravel()]
grid_df = pd.DataFrame(grid, columns=['Tinggi Badan (cm)', 'Berat Badan (kg)'])
Z = clf.predict(grid_df)
Z = Z.reshape(xx.shape)

plt.figure(figsize=(8, 6))
plt.contourf(xx, yy, Z, cmap=plt.cm.RdYlBu, alpha=0.6)

# Plot titik data training
for label, color in zip([0, 1], plot_colors):
    idx = np.where(y == label)
    plt.scatter(X.iloc[idx]['Tinggi Badan (cm)'], X.iloc[idx]['Berat Badan (kg)'],
                c=color, edgecolor='k', label=le.inverse_transform([label])[0])

# Plot data uji
plt.scatter(uji_baru['Tinggi Badan (cm)'], uji_baru['Berat Badan (kg)'],
            c='black', marker='X', s=100, label='Data Uji')

plt.xlabel("Tinggi Badan (cm)")
plt.ylabel("Berat Badan (kg)")
plt.title("Decision Boundary (Panjang vs Berat)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Visualisasi Decision Tree dalam bentuk gambar

```python
# === Visualisasi Struktur Pohon ===
plt.figure(figsize=(12, 3))
plot_tree(
    clf,
    feature_names=["Tinggi Badan (cm)", "Berat Badan (kg)"],
    class_names=le.classes_,  # gunakan label string hasil encoding
    filled=True,
    rounded=True,
    impurity=True  # tampilkan nilai Gini
)
plt.title("Decision Tree untuk Homo Sapiens vs Homo Erectus")
plt.tight_layout()
plt.show()
```
